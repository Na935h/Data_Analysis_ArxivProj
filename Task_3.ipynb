## Explanation for Task 3

In this task, I analyzed the number of authors for each paper in the arxiv metadata dataset collected in Task 1.

First, I loaded the `arxiv_papers_2025.csv` file using pandas. The dataset already contains an `author_count` column, which represents the number of authors for each paper.

I then grouped the papers into four categories based on the number of authors:

- Single author papers
- Two author papers
- Papers with 3 to 4 authors
- Papers with more than 4 authors

After calculating the number of papers in each category, I created a pie chart using Matplotlib. The pie chart shows the percentage distribution of papers in each author-count category. This helps visualize whether most papers in the sample were written by one author, two authors, about 3-4 authors, or more than 4 authors.

```python
# Task 3 - Creating a Pie Chart of Author Count Distribution
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sample_data/arxiv_papers_2025.csv")

print(f"Total papers: {len(df)}")
print(f"Sample author counts:\n{df['author_count'].value_counts().head(10)}")
```

```python
# Categorizing the papers by number of authors

single         = (df["author_count"] == 1).sum()
two            = (df["author_count"] == 2).sum()
three_four     = (df["author_count"].between(3, 4)).sum()
more_than_four = (df["author_count"] > 4).sum()

# Checking each paper written by how many number of authors
print(f"Single author:       {single}")
print(f"Two authors:         {two}")
print(f"3-4 authors:         {three_four}")
print(f"More than 4 authors: {more_than_four}")
print(f"Total:               {single + two + three_four + more_than_four}")
```

```python
# Generating the the Pie Chart

labels  = ["Single Author", "Two Authors", "3–4 Authors", "More than 4 Authors"]
sizes   = [single, two, three_four, more_than_four]
colors  = ["blue", "green", "red", "purple"]
explode = (0.05, 0.05, 0.05, 0.05)

fig, ax = plt.subplots(figsize=(8, 8))

wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct="%1.1f%%",
    startangle=140,
    textprops={"fontsize": 12}
)

ax.set_title("Distribution of Arxiv Papers by Number of Authors (2025)",
             fontsize=14, fontweight="bold", pad=20)

plt.tight_layout()
plt.show()
```
