## Explanation for Task 2

In this task, I used the metadata file created in Task 1 to create separate DataFrames for each quarter of the year 2025.

First, I loaded the CSV file `arxiv_papers_2025.csv` using pandas. Since the arxiv ID contains the year and month information, I extracted the month from each `arxiv_id`.

After extracting the month, I divided the dataset into four quarterly DataFrames:

- Q1: January to March
- Q2: April to June
- Q3: July to September
- Q4: October to December

Then, I printed the number of papers in each quarter and checked that the total number of papers across all four quarters matched the original dataset. I also displayed the first five rows of each quarterly DataFrame to verify that the data was separated correctly.

```python
# Task 2: Generating DataFrames for each quarter
import pandas as pd

df = pd.read_csv("/content/sample_data/arxiv_papers_2025.csv")

# Extracting the month from arxiv_id
df["arxiv_month"] = df["arxiv_id"].astype(str).apply(lambda x: int(x[2:4]))

print(f"Total papers loaded: {len(df)}")
print(f"\nPapers per month (from arxiv_id):")
print(df["arxiv_month"].value_counts().sort_index())
df.head()
```

```python
# Q1: January – March
df_q1 = df[df["arxiv_month"].between(1, 3)].reset_index(drop=True)

# Q2: April – June
df_q2 = df[df["arxiv_month"].between(4, 6)].reset_index(drop=True)

# Q3: July – September
df_q3 = df[df["arxiv_month"].between(7, 9)].reset_index(drop=True)

# Q4: October – December
df_q4 = df[df["arxiv_month"].between(10, 12)].reset_index(drop=True)

print(f"Q1 (Jan–Mar): {len(df_q1)} papers")
print(f"Q2 (Apr–Jun): {len(df_q2)} papers")
print(f"Q3 (Jul–Sep): {len(df_q3)} papers")
print(f"Q4 (Oct–Dec): {len(df_q4)} papers")
print(f"Total:        {len(df_q1) + len(df_q2) + len(df_q3) + len(df_q4)} papers")
```
